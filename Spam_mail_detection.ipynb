```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and preprocessing

```python
raw_mail_data = pd.read_csv("/content/sample_data/mail_data.csv")
```

Label Encoding

spam --> 0
nonspam --> 1

```python
raw_mail_data.loc[raw_mail_data['Category']=='spam','Category',]=0
raw_mail_data.loc[raw_mail_data['Category']=='ham','Category',]=1
```

seperate feature and target

```python
x = raw_mail_data['Message']
y = raw_mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

Train test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
```

Feature extraction

```python
# transform the text data to feature vectors that can be used as input for the logistic regresion model
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features =feature_extraction.transform(x_test)

# convert y_train and y_test values as integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train_features)
```

Model training

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

model evaluation

```python
# prediction on training data
prediction_train = model.predict(x_train_features)
accuracy_score_train = accuracy_score(y_train,prediction_train)
print("Training accuracy: ",accuracy_score_train)
```

```python
# prediction on test data
prediction_test = model.predict(x_test_features)
accuracy_score_test = accuracy_score(y_test,prediction_test)
print("Test accuracy: ",accuracy_score_test)
```

Building a predictive system

```python
mail = ["Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat..."]
input_data_features = feature_extraction.transform(mail)
prediction = model.predict(input_data_features)
if(prediction[0]==0):
  print("Spam")
else:
  print("Ham")
```
