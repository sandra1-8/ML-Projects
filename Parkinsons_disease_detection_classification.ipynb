```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinsons_data = pd.read_csv("/content/sample_data/parkinsons.csv")
```

```python
# distribution of target variable
parkinsons_data['status'].value_counts()
```

0 ---> people without Parkinsons
1 ---> people with Parkinsons

```python
# grouping the data based on target variable
parkinsons_data.groupby('status').mean(numeric_only=True)
```

Data Preprocessing

```python
# seperating features and target
x = parkinsons_data.drop(columns=['name','status'], axis=1)
y = parkinsons_data['status']

print(x)
print(y)
```

Train test split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)

x_test = scaler.transform(x_test)

print(x_train)
```

Model Training

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
```

```python
# training svm model with training data
model.fit(x_train,y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train,x_train_prediction)
print("Accuracy score of training data: ",training_data_accuracy)
```

```python
# accuracy score on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test,x_test_prediction)
print("Accuracy score of test data: ",test_data_accuracy)
```

Building a predictive system

```python
input_data = (162.56800,198.34600,77.63000,0.00502,0.00003,0.00280,0.00253,0.00841,0.01791,0.16800,0.00793,0.01057,0.01799,0.02380,0.01170,25.67800,0.427785,0.723797,-6.635729,0.209866,1.957961,0.135242)

input_data_as_array = np.asarray(input_data)
input_data_reshaped = input_data_as_array.reshape(1,-1)
# standardise the data
std_data = scaler.transform(input_data_reshaped)
prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 1):
  print("The person has Parkinson's disease")
else:
  print("The person does not have Parkinson's disease")
```
